```python
#Task 2
#Importing necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 


#Loading the tables 
df_movies=pd.read_csv('movies.csv')
df_credits=pd.read_csv('credits.csv')
df_keywords=pd.read_csv('keywords.csv')
df_ratings=pd.read_csv('ratings.csv')
```

```python
#Droping nan values from'budget' and ' revenue' columns of df_movies data frame. 
df_movies_n=df_movies.dropna(subset=['budget'])
df_movies_n1=df_movies_n.dropna(subset=['revenue'])
```

```python
#Plotting scatter plot of 'budget' vs. 'revenue'
fig,ax=plt.subplots()
ax.scatter(np.log10(df_movies_n1['budget']), np.log10(df_movies_n1['revenue']), color= 'magenta', alpha=0.2)

#Setting label for x and y axis
ax.set_xlabel('budget')
ax.set_ylabel('revenue')

#Setting tittle for graph
ax.set_title('Plot showing relation between budget and revenue')

plt.show()
```

```python
#Statistical test to determine significant correlation between revenue and budget

## Import the desired function from scipy.stats: 
from scipy.stats import spearmanr 

#Calculate spearman corelation coefficient and determine significance :
print(spearmanr(np.log10(df_movies_n1['budget']),np.log10(df_movies_n1['revenue'])))
```

```python
#Best fit relation
fig,ax=plt.subplots()
ax.scatter(np.log10(df_movies_n1['budget']), np.log10(df_movies_n1['revenue']), color= 'magenta', alpha=0.2)

#Setting label for x and y axis
ax.set_xlabel('budget')
ax.set_ylabel('revenue')

#Setting tittle for graph
ax.set_title('Plot showing relation between budget and revenue')

## The polyfit function takes an x array, a y array, and a degree: first degree = linear, second degree = quadratic, etc.

f=np.polyfit(np.log10(df_movies_n1['budget']),np.log10(df_movies_n1['revenue']),deg=1)
x_arr=np.linspace(0,8, 100)

y_mod=f[0]*x_arr+ f[1]

ax.plot(x_arr,y_mod, color='r')
plt.show()

```

```python
linear_prediction=f[0]*np.log10(df_movies_n1['budget'])+f[1]
                        
mse=np.mean((linear_prediction - np.log10(df_movies_n1['revenue']))**2)
print(mse)
```
