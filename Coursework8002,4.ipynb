```python
#Importing necessary library functions: 
import numpy as np
import pandas as pd
import json
import ast
#Importing CSV files 
df_keywords=pd.read_csv('keywords.csv')
df_movies=pd.read_csv('movies.csv')
df_credits=pd.read_csv('credits.csv')
df_ratings=pd.read_csv('ratings.csv')
```

```python
# remove rows with NaN values
df_kn= df_keywords.dropna(subset=['keywords'])

df_kn.head()
```

```python
#The ast.literal_eval function is used to safely evaluate the string representation of the list of dictionaries in each row. Then, a list comprehension is used to extract the values of the 'name' key from each dictionary. The results are stored in a new column called 'name_values'.

df_kn['keywords'] = df_kn['keywords'].apply(lambda x: [item['name'] for item in ast.literal_eval(x)])


# Print the modified DataFrame
df_kn.head(10)
```

```python
#Selecting columns named id,title, vote_count,vote_average that will be needed to search for keywords related to the plots of movies
mn=['id','title','vote_count','vote_average']
df_mn=df_movies[mn] 
df_total=pd.merge(df_kn,df_mn,on='id',how='inner')
df_total

```

```python
#Selecting a sensible default minimum vote count,since mode=1, mean= 109, max= 104075, min= 0
vc= 109 
```

```python
#A function that allows users to input and recommend the top-rated movies that match those keywords
in1=input("Enter your desired type:")
in2=input('Enter your minimun vote count:')
def movie_recommendation(a,b):
    search_string = a 
    b = int(b)
    matches = df_total[df_total['keywords'].apply(lambda x: search_string in x)]
    filter_df = matches[(matches['vote_count'] >= b) & (matches['vote_average'] > 6)]
    name_mov = filter_df['title']
    return name_mov
movie_recommendation(in1, in2)
```
